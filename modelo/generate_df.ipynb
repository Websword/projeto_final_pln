```python
# Função para gerar frases
def gerar_frases(prefixos, termos, comandos, repeticoes=10):
    exemplos = []
    target_exemplos = []
    for _ in range(repeticoes):
        for prefixo in prefixos:
            for termo in termos:
                exemplos.append(f"{prefixo} {termo} ")
                target_exemplos.append(comandos[termos.index(termo)])
                print(f"{prefixo} {termo} ---> {comandos[termos.index(termo)]}")
    return exemplos, target_exemplos


# Gerando frases para ações (turbo e pausa)
def gerar_frases_acoes(prefixos, termos, comandos, repeticoes=10):
        
    exemplos_acoes = []
    target_acoes = []
    for acao, variacoes in acoes_variacoes.items():
        for prefixo in prefixo_acoes:
            for variacao in variacoes:
                for _ in range(5):  # Diversidade de cada variação
                    exemplos_acoes.append(f"{prefixo} {variacao}")
                    target_acoes.append(acao)
                    print(f"{prefixo} {variacao} ---> {acao}")
    return exemplos_acoes, target_acoes

```

```python
# Gerando frases para movimentos
exemplos_moves,exemplos_target = gerar_frases(prefixo_moves, moves, game_keys_moves, repeticoes=15)
exemplos_acoes,acoes_target = gerar_frases_acoes(prefixo_acoes, game_keys_acoes, acoes_variacoes, repeticoes=10)

# Juntando tudo
exemplos = exemplos_moves + exemplos_acoes
target = exemplos_target + acoes_target


for i in range(len(exemplos)):
    print(f"{exemplos[i]} ---> {target[i]}")


df = pd.DataFrame()
df['comando'] = exemplos
df['target'] = target





# Salvando o DataFrame em um arquivo CSV
df.to_csv('../dataset/dataset_comandos.csv', index=False)
```
